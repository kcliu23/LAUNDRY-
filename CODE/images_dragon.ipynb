```python
import csv
import pandas as pd
import numpy as np
import matplotlib
from  matplotlib import pyplot as plt
from sklearn import preprocessing
from BaselineRemoval import BaselineRemoval
%matplotlib inline
```

```python
data_arr = []
data_arry = []
```

```python
count = 0
for i in range(1,13):
    with open(f'../data/1031_kc_nano1_result_{i}.csv','rt') as csvfile:
        rows = csv.reader(csvfile)
        df=pd.read_csv(f'../data/1031_kc_nano1_result_{i}.csv')
        for row in rows:
              data_arr.append(row)
        for i in range(0, len(df)+1):
              data_arry.append(float(data_arr[i+count][4]))
        print(len(data_arry))
        count +=300000
```

```python
polynomial_degree=2 #only needed for Modpoly and IModPoly algorithm
baseObj=BaselineRemoval(data_arry)        
y=baseObj.IModPoly(polynomial_degree)
```

```python
y = y /2.5
```

```python
plt.plot(y,color = '#3F9475',linewidth = 0.5)
plt.savefig(f'../images/1031A_{i}')
```

```python
px = 1/plt.rcParams['figure.dpi']  
for i in range(int(len(y)/100-500)):
    plt.figure(figsize = (224.5*px,224.5*px))
    plt.plot(y[100*i:100*i+499],color = '#3F9475',linewidth = 1)
    plt.ylim(-1,1)
    plt.xlabel(f'T_{i}')
    plt.tight_layout()
    plt.savefig(f'../images/1031A_{i}')
```
